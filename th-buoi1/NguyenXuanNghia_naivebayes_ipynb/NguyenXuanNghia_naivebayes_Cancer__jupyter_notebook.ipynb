# Import Library

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

# Read data datacum.txt and split it to train data and test data


```python
def read_data(data_fn):
    with open(data_fn) as f:
        content = f.readlines()
    content = [x.strip() for x in content]

    data = [[int(y) for y in x.split(',')]
            for x in content if x and '#' not in x]
    benigns = 80
    malignant = 40
    test_data = [0] * (benigns + malignant)
    index = 0

    # split test data out of data and not get code attribute
    for ele in data:
        if (ele[1] == 2 and benigns > 0):
            test_data[index] = ele[1:]
            index += 1
            benigns -= 1
        if (ele[1] == 4 and malignant > 0):
            test_data[index] = ele[1:]
            index += 1
            malignant -= 1
        if (index == len(test_data)):
            break
    
    # test label has type attribute of core and remove type attribute out of test_data
    test_label = [ele.pop(0) for ele in test_data]

    # train data not get code attribute of core and type attribute
    train_data = [ele[1:] for ele in data if ele not in test_data]
    
    # train label has type attribute of core and remove type attribute out of train_data
    train_label = [ele.pop(0) for ele in train_data]

    return test_data, test_label, train_data, train_label

path = "./Cancel_Data/NB_last/datacum.txt"
test_data, test_label, train_data, train_label = read_data(path)

print("\n test-data : ",test_data)
print("\n test-label : ",test_label)
print("\n train-data : ",train_data)
print("\n train-label",train_label)
```

# Fitting model


```python
model_sk = GaussianNB()
model_sk.fit(train_data, train_label)
```

# Predict test data


```python
y_pred = model_sk.predict(test_data)
```

# Confusion matrix and evaluation metrics for test data

```python
conf_matrix = confusion_matrix(y_true=test_label, y_pred=y_pred)
accuracy = accuracy_score(np.array(test_label), y_pred)
precision = precision_score(test_label, y_pred, average=None)
recall = recall_score(test_label, y_pred, average=None)

print("confusion matrix:\n",conf_matrix)
print("accuracy:", accuracy)
print("precision:", precision)
print("recall:", recall)
```
