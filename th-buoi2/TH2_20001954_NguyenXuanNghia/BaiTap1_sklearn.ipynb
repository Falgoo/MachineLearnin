## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

## Read data

```python
data = pd.read_csv("./SAT_GPA.csv")
data.describe()
y = data['GPA']
x = data['SAT']

plt.scatter(x,y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

## Split data to train data and validate data

```python
x = np.asarray(x).reshape(-1,1)
y = np.asarray(y).reshape(-1,1)

train_data = x[:60]
y_data = y[:60]

validate_data = x[60:]
y_validate = y[60:]
```

## Fitting model

```python
regr = linear_model.LinearRegression()
regr.fit(train_data, y_data)
```

## Predict validate model of train data

```python
y_pre = regr.predict(train_data)
print("y predict:",y_pre.T)
```

## Draw model

```python
plt.scatter(train_data, y_data)
fig = plt.plot(train_data, y_pre, lw=4, c='orange', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

## Predict validate model of validate data and show mean square error

```python
y_pre = regr.predict(validate_data)
mse = mean_squared_error(y_validate, y_pre)

print("validate predict:\n",y_pre.T)
print("\nMean square error is:",mse)
```
